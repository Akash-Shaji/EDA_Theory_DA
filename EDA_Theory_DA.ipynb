```python
# Splitting the datafrane based on diagnosis classes. Each class will be processed seperately and merged

df_M = df[df['diagnosis'] == 'M']
df_B = df[df['diagnosis'] == 'B']
```

```python
# Describe Seperately

print("For M")
print(df_M.describe())
print("For B")
print(df_B.describe())
```

```python
# Identifying if any cells have value set as 0 and replacing them with respective mean

for col in df_M.columns:
  if df_M[col].dtype == 'object':
    continue
  else:
    if (df_M[col] == 0).sum() > 0:
      mean = df_M[df_M[col]!=0][col].mean()
      df_M[col] = df_M[col].replace(0, df_M[col].mean())

for col in df_B.columns:
  if df_B[col].dtype == 'object':
    continue
  else:
    if (df_B[col] == 0).sum() > 0:
      mean = df_B[df_B[col]!=0][col].mean()
      df_B[col] = df_B[col].replace(0, df_B[col].mean())
```

```python
print("For M")
print(df_M.describe())
print("For B")
print(df_B.describe())
```

```python
#Sunray plot

stats = df.groupby('diagnosis')['radius_worst'].agg(['mean', 'median', 'min', 'max', 'std'])
labels = stats.columns
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

def add_sunray(diagnosis, color):
    values = stats.loc[diagnosis].tolist()
    values += values[:1]

    ax.fill(angles, values, color=color, alpha=0.25)
    ax.plot(angles, values, color=color, linewidth=2, label=f'{diagnosis} Diagnosis')

add_sunray('M', 'red')
add_sunray('B', 'blue')

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.title('Sunray Plot of Radius_worst Statistics Coloured by Diagnosis - 21BCE2348', size=14, y=1.3)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```

# Clustering - Part 2 of assignment

```python
#Kmeans Clustering

#Taking only 2 features for ease of visualization

X = df.iloc[:, [2, 3]].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method - 21BCE2348')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#From the graph we can see that the optimal number of clusters is 2
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Patients')
plt.xlabel('radius_mean')
plt.ylabel('texture_mean')
plt.legend()
plt.show()
```

```python
#All numeric_columns
numerical_columns = df.iloc[:, 2:]
```

```python
#Hierarchial Clustering

import scipy.cluster.hierarchy as sch

plt.figure(figsize=(15,6))
plt.title('Dendrogram')
plt.xlabel('Patients')
plt.ylabel('Euclidean distances')

dendrogram = sch.dendrogram(sch.linkage(numerical_columns, method = 'ward'))
plt.show()
```
